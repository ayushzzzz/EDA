```python
def convert_size(value):
    if 'M' in value:
        return float(value.replace('M', '')) * 1_000_000  
    elif 'k' in value:
        return float(value.replace('k', '').replace(',', '')) * 1_000  
    elif '+' in value or ',' in value: 
        return float(value.replace('+', '').replace(',', ''))  
    elif value == 'Varies with device':
        return np.nan  
    else:
        return float(value)  


df_copy['Size'] = df_copy['Size'].apply(convert_size).astype(float)
```

```python
df_copy['Size'] = df_copy['Size'].fillna(0).astype(float).astype(int)
print(df_copy[df_copy['Size'] == 0])

```
