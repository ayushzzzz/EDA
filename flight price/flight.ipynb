```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_excel('flight_price (1).xlsx')
```

```python
flights.head()
```

```python
flights['date'] = flights['Date_of_Journey'].str.split('/').str[0].astype(int)
flights['month'] = flights['Date_of_Journey'].str.split('/').str[1].astype(int)
flights['year'] = flights['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
flights.info()
```

```python
flights.isna().value_counts()
```

```python
flights.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
flights.head()
```

```python
flights['Arrival_Time'] = flights['Arrival_Time'].str.split(' ').str[0]
flights['Arrival_Time']
```

```python
flights.head()
```

```python
flights['Arrival_hour']=flights['Arrival_Time'].str.split(':').str[0].astype(int)
flights['Arrival_min']=flights['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
flights.drop('Arrival_Time',axis=1,inplace=True)
```

```python
flights['Departure_hour']=flights['Dep_Time'].str.split(':').str[0].astype(int)
flights['Departure_min']=flights['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
flights.drop('Dep_Time',axis=1,inplace=True)
```

```python
flights.info()
```
