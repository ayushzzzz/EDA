```python
#correlation
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(red_wine.corr(),annot=True)
```

```python
red_wine['quality'].value_counts().plot(kind='bar')
```

```python
for column in red_wine.columns:
    sns.histplot(red_wine[column],kde=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the number of rows and columns for subplots
num_cols = len(red_wine.columns)
fig, axes = plt.subplots(nrows=num_cols, figsize=(8, num_cols * 3))

# Plot each column separately
for i, column in enumerate(red_wine.columns):
    sns.histplot(red_wine[column],kde=True,ax=axes[i])

plt.tight_layout()
plt.show()
```

```python
sns.catplot(x='quality',y='fixed acidity',data=red_wine,kind='box')
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='alcohol',y='fixed acidity',hue='quality',data=red_wine)
```
